```python
from transformers import pipeline

sentiment = pipeline("sentiment-analysis")

texts = [
    "I love this phone. The battery life is excellent!",
    "Worst purchase I ever made."
]

results = sentiment(texts)
for t, r in zip(texts, results):
    print(t, "->", r)
```

```python
from sentence_transformers import SentenceTransformer, util

model = SentenceTransformer('all-MiniLM-L6-v2')

sentences = [
    "How do I cook pasta?",
    "Best ways to boil spaghetti?",
    "What is the capital of France?"
]

embeds = model.encode(sentences, convert_to_tensor=True)
print(embeds.shape)
```

```python
query = "How to make spaghetti?"
q_emb = model.encode(query, convert_to_tensor=True)

scores = util.pytorch_cos_sim(q_emb, embeds)[0]

print("Query:", query)
for s, score in zip(sentences, scores):
    print(f"{score.item():.3f}\t{s}")
```

```python
from datasets import load_dataset
ds = load_dataset("imdb")
print(ds)
```

```python
small_train = ds['train'].shuffle(seed=42).select(range(4000))
small_test  = ds['test'].shuffle(seed=42).select(range(2000))
```

```python
import pandas as pd

train_df = pd.DataFrame(small_train)
test_df = pd.DataFrame(small_test)

train_df.head()
```

```python
from fastai.text.all import *

dls = TextDataLoaders.from_df(
    train_df,
    text_col='text',
    label_col='label',
    valid_pct=0.1,
    bs=32
)
```

```python
from fastai.text.all import *
from fastprogress.fastprogress import master_bar, progress_bar

dls = TextDataLoaders.from_df(
    train_df,
    text_col='text',
    label_col='label',
    valid_pct=0.1,
    bs=32
)

learn = text_classifier_learner(dls, AWD_LSTM, drop_mult=0.5, metrics=accuracy)
with learn.no_bar(): # Temporarily disable progress bar
    learn.fine_tune(1)
```

```python
import csv

texts = [
    "I love this phone!",
    "Terrible product!"
]

results = pipeline("sentiment-analysis")(texts)

with open("sentiment_results.csv", "w", newline="") as f:
    writer = csv.writer(f)
    writer.writerow(["text","label","score"])
    for t, r in zip(texts, results):
        writer.writerow([t, r['label'], r['score']])

print("Saved sentiment_results.csv")
```
